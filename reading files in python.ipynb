```python
import json

def read_covid(filename):
    file = open(filename)
    list_data = json.load(file)
    cases = []
    for data in list_data:
        cases.append(data['Cases'])
    return cases

out_dict = {}

out_dict['indonesia'] = read_covid('docs/covid indonesia.json')
out_dict['jepang'] = read_covid('docs/covid japan.json')
out_dict['india'] = read_covid('docs/covid india.json')

out_dict
```
